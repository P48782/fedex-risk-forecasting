# 04 - Prescriptive Modeling
Suggest buffering, re-routing, or rebalancing strategies.

```python
import pandas as pd

df = pd.read_csv('../data/processed/fedex_cleaned.csv')
high_risk = pd.read_csv('../data/processed/route_risk_scores.csv')

# Suggest buffer of 1 day for high risk routes
high_risk_routes = list(zip(high_risk['Source'], high_risk['Destination']))
df['Recommended_Buffer'] = df.apply(
    lambda row: 1 if (row['Source'], row['Destination']) in high_risk_routes else 0,
    axis=1
)

# Export recommendations
prescriptive_output = df[['Source', 'Destination', 'Delay', 'Recommended_Buffer']].drop_duplicates()
prescriptive_output.to_csv('../data/processed/prescriptive_actions.csv', index=False)

prescriptive_output.head(10)
```
